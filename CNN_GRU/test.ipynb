```python
import numpy as np
import cPickle as pickle
from keras.models import model_from_json
```

```python
json_file = open('zippyATIS/model_architecture.json', 'r').read()
model = model_from_json(json_file)
model.load_weights("zippyATIS/best_model_weights.h5")
```

```python
train_set, valid_set, dicts = pickle.load(open('/home/vishal/Documents/ATIS.keras/atis.pkl', 'rb'))
w2idx, labels2idx = dicts['words2idx'], dicts['labels2idx']
train_x, _, train_label = train_set

idx2w  = {w2idx[k]:k for k in w2idx}
idx2la = {labels2idx[k]:k for k in labels2idx}
```

```python
def to_vector(sentence, w2idx):
    vec = []
    for word in sentence.split():
        idx = w2idx.get(word)
        if idx == None:
            idx = w2idx['<UNK>']
        vec.append(idx)
    return vec
```
